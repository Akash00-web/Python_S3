```python
from functools import *
l=[1,2,3,4]
x1=list(map(lambda i: i*i,l))
print(x1)
x2=list(filter(lambda i:i%2==1,l))
print(x2)
x3=reduce(lambda i,y:i*y,l)
print(x3)
```

```python
x=[4,4,4,1,1,1,1,2,2,3,5,5,5,5,5,6]
d={}
for i in x:
    d[i]=d.get(i,[])
    d[i].append(i)
z=sorted(d.items(), key=lambda x:len(x[1]))
print(z)
```

```python
x=[4,4,4,1,1,1,1,2,2,3,5,5,5,5,5,6]
d={}
for i in x:
    d[i]=d.get(i,[])
    d[i].append(i)
z=sorted(d.items(), key=lambda x:x[1][0])
print(z)
```
