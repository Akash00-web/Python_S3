```python
op=""
x=input('Enter string')
key=int(input("Enter jump to"))
x=x.upper()
ans="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in x:
    if i in ans:
        sol=(ans.find(i)+key)%26
        op=op+ans[sol]
    else:
        op=op+i
print(op)
```
