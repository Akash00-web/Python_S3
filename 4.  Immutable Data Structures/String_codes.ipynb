```python
#Write a Python function that accepts a string and calculate the number of uppercase letters and lowercase letters.
def abc(s):
    u=0
    l=0
    for i in s:
        if i.isupper():
            u=u+1
        if i.islower():
            l=l+1
    return u,l

u,l=abc("Aakash")
print(f" upper {u} and lower {l}")
```

```python
#Write a program to calculate the sum and average of the digits present in a string.

s=input("Enter string ")
ans=0
count=0
for i in s:
  
    if i.isdigit():
        ans=ans+int(i)
        count+=1
print(f"""sum of digits {ans} 
   avg of digits {ans/count}
     """)
        
```

```python
#Write a Python program to print even length words in a string.
s=input("Enter string")
s=s.split()
for i in s:
    if len(i)%2==0:
        print(i,end=" ")
```

```python
# Write a program to check if two strings are balanced. For example, strings s1 and s2 are balanced if all the 
#characters in the s1 are present in s2. The character’s position doesn’t matter.

s1='ABCDEFG'
s2='GFABCDE'
def abc(s1,s2):
    for i in s1:
        if i not in s2:
            return False
    return True
if abc(s1,s2):
    print("BAlanced")
else:
    print("Not balanced")
    
```
